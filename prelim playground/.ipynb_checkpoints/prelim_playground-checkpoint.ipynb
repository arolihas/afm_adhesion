# Physical Data from AFM

```python
import pandas
from os import listdir, getcwd
from os.path import isfile, join
```

```python
path = getcwd()
files = [f for f in listdir(path) if isfile(join(path, f)) and '.csv' in f]
files
```

```python
data_list = {}
for f in files:
    data_list[f] = pandas.read_csv(f)
```

```python
for f in files:
    print(data_list[f].columns)
```

```python
data_list[files[0]] = data_list[files[0]].drop([' No'], axis=1)
```

```python
data_list[files[0]] = data_list[files[0]].rename(columns={'ZIndex':'No'})
for f in files:
    data_list[f] = data_list[f].reindex(columns=['No','zSensr','defl'])
```

```python
approach_curve = []
approach_constants = []
full_curve = []
full_constants = []
for k in data_list:
    if 'Apr' in k:
        approach_curve.append(data_list[k])
        approach_constants.append(float(k[-9:-4]))
    else:
        full_curve.append(data_list[k])
        full_constants.append(float(k[-9:-4]))
for i in range(len(approach_curve)):
    approach_curve[i].defl *= approach_constants[i]
```

**Visualizations**

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['agg.path.chunksize'] = 10000

d = approach_curve[0].defl * approach_constants[0]
z = approach_curve[0].zSensr
t = approach_curve[0].No

plt.figure(1)
plt.suptitle('Approach Curve 0')
plt.subplot(211)
plt.plot(t, d)
plt.ylabel('Force (N)')
plt.subplot(212)
plt.plot(t, z)
plt.ylabel('Z position')
```

```python
matplotlib.rcParams['agg.path.chunksize'] = 10000
n = 1
d = approach_curve[n].defl * approach_constants[n]
z = approach_curve[n].zSensr
t = approach_curve[n].No

plt.figure(1)
plt.suptitle('Approach Curve ' + str(n))
plt.subplot(211)
plt.plot(t, d)
plt.ylabel('Force (N)')
plt.subplot(212)
plt.plot(t, z)
plt.ylabel('Z position')
```

# K-Means Analysis

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler 
```

**Approach Curve 0**

```python
n = 0
X = approach_curve[n].drop(['zSensr'], axis=1).values
scaler = StandardScaler()  
scaler.fit_transform(X)   
```

**Elbow Method**

```python
inertia_scores = []
for i in range(1,10):
    Xk = X[:]
    km = KMeans(i, random_state=10)
    km.fit_transform(Xk)
    labels = km.predict(Xk)
    inertia_scores.append(km.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1,10),inertia_scores)
plt.xlabel("Clusters")
plt.ylabel("Inertia")
plt.title("Finding Optimal K through Elbow Method")
plt.grid(True)
```

**Gap Statistic Method**

```python
import numpy as np
from gap_statistic import OptimalK
optimalK = OptimalK(parallel_backend='multiprocessing')
n_clusters = optimalK(X, cluster_array=np.arange(1, 10))
print('Optimal clusters: ', n_clusters)
```

```python
plt.plot(optimalK.gap_df.n_clusters, optimalK.gap_df.gap_value, linewidth=3)
plt.scatter(optimalK.gap_df[optimalK.gap_df.n_clusters == n_clusters].n_clusters,
            optimalK.gap_df[optimalK.gap_df.n_clusters == n_clusters].gap_value, s=250, c='r')
plt.grid(True)
plt.xlabel('Cluster Count')
plt.ylabel('Gap Value')
plt.title('Gap Values by Cluster Count')
plt.show()
```

**Labeling**

```python
Xk = X[:]
km = KMeans(3, random_state=10)
labels = km.fit_predict(Xk)
```

```python
plt.scatter(approach_curve[n].No, approach_curve[n].defl, c=labels)
plt.ylim(min(Xk[:,1]),max(Xk[:,1]))
plt.title('3-Clustering of Approach Curve {}'.format(n))
plt.ylabel('Force (N)')
```

```python
plt.figure(1)
plt.suptitle('3-Clustering Approach Curve 0')
plt.subplot(211)
plt.scatter(range(Xk.shape[0]), Xk[:,0], c=labels)
plt.ylim(min(Xk[:,0]),max(Xk[:,0]))
plt.ylabel('Force (N)')
plt.subplot(212)
plt.scatter(range(Xk.shape[0]), Xk[:,1], c=labels)
plt.ylim(min(Xk[:,1]),max(Xk[:,1]))
plt.ylabel('Z position')
```

```python
Xkn = X[:]
km = KMeans(9, random_state=10)
labelsn = km.fit_predict(Xkn)
```

```python
plt.scatter(approach_curve[n].No, approach_curve[n].defl, c=labelsn)
plt.ylim(min(Xkn[:,1]),max(Xkn[:,1]))
plt.title('9-Clustering of Approach Curve {}'.format(n))
plt.ylabel('Force (N)')
```
