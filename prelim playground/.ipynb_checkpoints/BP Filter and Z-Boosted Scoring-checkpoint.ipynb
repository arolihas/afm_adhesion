**Peak Detection Algorithm**

*Algorithm from https://stackoverflow.com/a/22640362*

*Implementation from https://gist.github.com/dlegor/286dd5cd5ecc317c34ed81aaed4a2b37*

```python
from numba.decorators import jit
import numpy as np
import pylab
#The original version is here: https://gist.github.com/ximeg/587011a65d05f067a29ce9c22894d1d2
#I made small changes and used numba to do it faster.
  
@jit
def thresholding_algo2(y, lag, threshold, influence):
    signals = np.zeros(len(y))
    filteredY = np.array(y)
    avgFilter = np.zeros(len(y))
    stdFilter = np.zeros(len(y))
    avgFilter[lag - 1] = np.mean(y[0:lag])
    stdFilter[lag - 1] = np.std(y[0:lag])
    for i in range(lag, len(y) - 1):
        if abs(y[i] - avgFilter[i-1]) > threshold * stdFilter [i-1]:
            if y[i] > avgFilter[i-1]:
                signals[i] = 1
            else:
                signals[i] = -1

            filteredY[i] = influence * y[i] + (1 - influence) * filteredY[i-1]
            avgFilter[i] = np.mean(filteredY[(i-lag):i])
            stdFilter[i] = np.std(filteredY[(i-lag):i])
        else:
            signals[i] = 0
            filteredY[i] = y[i]
            avgFilter[i] = np.mean(filteredY[(i-lag):i])
            stdFilter[i] = np.std(filteredY[(i-lag):i])

    return dict(signals = np.asarray(signals),
                avgFilter = np.asarray(avgFilter),
                stdFilter = np.asarray(stdFilter))
```

**Data Collection and Cleaning**

```python
import pandas
from os import listdir, getcwd
from os.path import isfile, join
```

```python
path = getcwd()
files = [f for f in listdir(path) if isfile(join(path, f)) and '.csv' in f]
files
```

```python
data_list = {}
for f in files:
    data_list[f] = pandas.read_csv(f)
```

```python
data_list[files[0]] = data_list[files[0]].drop([' No'], axis=1)
```

```python
data_list[files[0]] = data_list[files[0]].rename(columns={'ZIndex':'No'})
for f in files:
    data_list[f] = data_list[f].reindex(columns=['No','zSensr','defl'])
```

```python
approach_curve = []
approach_constants = []
full_curve = []
full_constants = []
for k in data_list:
    if 'Apr' in k:
        approach_curve.append(data_list[k])
        approach_constants.append(float(k[-9:-4]))
    else:
        full_curve.append(data_list[k])
        full_constants.append(float(k[-9:-4]))
for i in range(len(approach_curve)):
    approach_curve[i].defl *= approach_constants[i]
```

**Algorithm**

```python
def peak_detection(force_curve, resonant_frequency):
    
    #Band pass filter at resonant frequency
    yf = fftp.fft(force_curve)
    bp = yf[:]
    for i in range(len(bp)):
        if i >= resonant_frequency:
            bp[i] = 0

    #Invert to smoothed curve, resampled and shifted
    ifft = fftp.ifft(yf)
    resample = ifft[::10000]
    shift = int(len(resample) / 10)
    y = resample[shift:-shift]
    
    #Sliding window algorithm detects peaks on  gradient of curve
    grad = np.gradient(y)
    sig_dict = thresholding_algo2(grad, int(len(grad)/5), 5, 0)
    
    #Retrieve the indices of where the cantilever bends
    indices = []
    bound = -1
    for i, v in enumerate(sig_dict['signals']):
        if v == -1 and i > bound:
            signal = [i + shift]
            while v == -1 and i <= len(sig_dict['signals']):
                i += 1
                bound = i
                v = sig_dict['signals'][i]
                if (v > -1):
                    continue
                signal.append(i + shift)
            indices.append(signal)
            
    #Force interactions at each deflection
    markers_on = []
    for signal in indices:
        pre_contact_average = np.real(np.mean(y[:signal[0]]))
        contact_force = force_curve[signal[-1] * 10000]
        force_interaction = abs(contact_force - pre_contact_average)
        print("Force: ", force_interaction)
        points = [signal[0] - 1, signal[-1]]
        markers_on.append(points)
    
    #Plot resampled with markers where window should be
    fig = plt.figure()
    ax=fig.add_axes([0,0,1,1])
    ax.plot(resample, marker='o', markevery=markers_on)
    #Plot rectangular window on original curve
    print(markers_on)
    fig = plt.figure()
    ax=fig.add_axes([0,0,1,1])
    ax.plot(force_curve)
    for signal in markers_on:
        ax.add_patch(
            patches.Rectangle((signal[0]*10000,force_curve[signal[0]*10000]),
                              (signal[1] - signal[0])*10000,
                              force_curve[signal[1]*10000] - force_curve[signal[0]*10000],
                              linewidth=1,
                              edgecolor='r',
                              facecolor='none',
                              zorder=4
                            )
        )
```
